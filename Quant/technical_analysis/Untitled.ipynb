```python
import talib
ubb, mbb, lbb = talib.BBANDS(aapl.Close, 20, 2) # 20일 기준선 , 표준편차 2 
#upperband , middleband, lowerband = BBAMDS(close, timeperiod=5, nbdevup=2, nbdevdn=2, maypte=0)


aapl['ubb'] = ubb
aapl['mbb'] = mbb
aapl['lbb'] = lbb
```

```python
aapl['ma5'] = aapl['Close'].rolling(window=5).mean() # 5일 이동평균 
aapl['ma60'] = aapl['Close'].rolling(window=60).mean() # 60일 이동평균 
```

```python
#볼린저 밴드 
#밴드의 폭이 이전보다 상대적으로 크거나 줄어들 경우, 과매수 또는 과매도 상태로가늠한다. 



# 볼린저 밴드 기반 트레이딩 북 
trading_book = aapl.copy().assign(action='',state='').reset_index() # assign 새로운열 할당 
trading_book.head(30)
```

```python
#문자열 길이 확인 
print(len(trading_book))
```

```python
#Trading 

for i in range(len(trading_book)):
    # 초기값 설정
    if i==0:
        # 종가가 하단밴드 이하이면 매수
        if trading_book.loc[i, 'Close'] <= trading_book.loc[i, 'lbb']:
            trading_book.loc[i, 'action'] = 'buy'
            trading_book.loc[i, 'state'] = 'holding'
        # 이외에는 아무것도 안하기
        else:
            trading_book.loc[i, 'action'] = ''
            trading_book.loc[i, 'state'] = ''
    else:
        # 종가가 하단밴드 이하인데
        if trading_book.loc[i, 'Close'] <= trading_book.loc[i, 'lbb']:
            # 이전상태가 보유중이 아니면 매수
            if trading_book.loc[i-1, 'state'] not in ['holding']:
                trading_book.loc[i, 'action'] = 'buy'
                trading_book.loc[i, 'state'] = 'holding'
            # 이전상태가 보유중이면 그대로 보유
            elif trading_book.loc[i-1, 'state'] in ['holding']:
                trading_book.loc[i, 'action'] = 'hold'
                trading_book.loc[i, 'state'] = 'holding'

        # 종가가 상단밴드 이상인데
        elif trading_book.loc[i, 'Close'] >= trading_book.loc[i, 'ubb']:
            # 이전상태가 보유중이 아니면 그대로 있기
            if trading_book.loc[i-1, 'state'] not in ['holding']:
                trading_book.loc[i, 'action'] = 'hold'
            # 이전상태가 보유중이면 매도
            elif trading_book.loc[i-1, 'state'] in ['holding']:
                trading_book.loc[i, 'action'] = 'sell'
                trading_book.loc[i, 'state'] = ''

        # 그 외에는 아무 행동도 하지 않고 상태는 이전상태 그대로 유지
        else:
            trading_book.loc[i, 'action'] = ''
            trading_book.loc[i, 'state'] = trading_book.loc[i-1]['state']
```

```python
report = trading_book.loc[trading_book['action'].isin(['buy', 'sell']), ['Date', 'Close', 'ubb', 'lbb', 'action', 'state']].reset_index(drop=True)
report
```

```python
report = report.assign(buy_price=report['Close'].shift(1)) # buy_price 한 행 밑으로 내리기 
report
```
