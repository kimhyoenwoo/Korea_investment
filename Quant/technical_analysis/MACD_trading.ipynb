```python
##########
### sample 
##########
df_data = fdr.DataReader('006840','2022-01-01')
df_data
```

```python
# Create macd index

# MACD 단기 
df_data['MACD_short'] = df_data['Close'].ewm(12).mean() # evm 지수이동평균 
# df_data.head()

# MACD 장기 
df_data['MACD_long'] = df_data['Close'].ewm(26).mean() #evm 지수이동평균 
# df_data.head()

# MACD 
df_data['MACD'] = df_data.apply(lambda x:(x['MACD_short'] -x['MACD_long']),axis=1)

# MACD Siginal
df_data['Siginal'] = df_data['MACD'].rolling(window=9).mean()

print(df_data)
'''매매기법
1.macd 곡선이 siginal 곡선을 상향 돌파 할 때 매수, 하향 돌파할떄 매도 
2.MACD 값이 0 선을 상향돌파시 매수, 하향돌파할때 매도 
3. MACD 값이 siginal선을 크로스 상향돌파할떄 매수시점, 크로스 하향돌파할떄를 매도시점으로 인식 
'''



```
